```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
columns_names=['user_id', 'item_id', 'rating', 'timestamp']
rate=pd.read_csv('udemy/u.data', sep='\t', names=columns_names)
rate.head()
```

```python
movie_titles=pd.read_csv('udemy/Movie_Id_Titles')
movie_titles.head()
```

```python
df=pd.merge(rate, movie_titles, on='item_id')
df.head()
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False) #series,
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
ratings['num of ratings']=pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
ratings.head()
```

```python
moviemat=df.pivot_table(index='user_id', columns='title', values='rating')
moviemat.head()
```

```python
starwars_user_ratings=moviemat['Star Wars (1977)']
```

```python
liarliar_user_ratings=moviemat['Liar Liar (1997)']
```

```python
similar_to_starwars=moviemat.corrwith(starwars_user_ratings)
```

```python
similar_to_liarliar=moviemat.corrwith(liarliar_user_ratings)
```

```python
corr_starwars=pd.DataFrame(similar_to_starwars, columns=['Correlation'])
corr_starwars.head()
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars=corr_starwars.join(ratings['num of ratings'])  #important to have sufficiemt number of rations
```

```python
corr_starwars=corr_starwars[corr_starwars['num of ratings']>100].sort_values('Correlation', ascending=False).head()
```
